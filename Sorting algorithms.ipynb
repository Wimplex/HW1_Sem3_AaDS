#Задача сортировки

**Примеры исопльзования сортировки**: использование в словарях, сортировка товаров по цене, сортировка игральных карт в руке игрока и т.д.

Задача сортировки описывается следующим образом:

* **Вход**: Последовательность $n$ чисел ($a_1, a_2, ..., a_n$).

* **Выход**: Перестановка ($a^{'}_1, a^{'}_2, ..., a^{'}_n$) исходной последовательности, для которой $a^{'}_1 \le a^{'}_2 \le ... \le a^{'}_n$ или $a^{'}_1 \ge a^{'}_2 \ge ... \ge a^{'}_n$ (в зависимости от требуемого порядка).

Например, получив на вход $\left \langle 31, 41, 59, 26, 41, 58\right \rangle$, алгоритм сортировки должен выдать на выход $\left \langle 26, 31, 41, 41, 58, 59\right \rangle$.

###Алгоритмы сортировки
* Сортировка методом пузырька (Bubble sort)

* Сортировка вставками (Insertion sort)

* Сортировка выборками (Selection sort)

* **Сортировка Шелла (Shell sort)**

* **Быстрая сортировка (Quick sort)**

* **Сортировка слиянием (Merge sort)**

Классификация алгоритмов сортировки:

* Временная сложность алгоритма (time complexity) - время затрачиваемое на сортировку всех элементов последовательности.

* Емкостная сложность алгоритма (space complexity) - занимает постоянное количество памаяти или требуемое количество памяти растет со ростом входа.

* Стабильность - означает эквивалентность ключей (элементов).

* Внутренняя сортировка (все хранится в RAM) или внешняя сортировка (все хранится на диске).

* Рекурсивный или нерекурсивный алгоритм.

````python
def generateRandomNumbers(n, start=0, stop=100):
    import random
    return [random.randint(start, stop) for r in range(n)]
````

````python
def inputIntegers(msg=''):
    return [int(n) for n in input(msg).split()]
````

####Алгоритм на псевдоязыке
```
BubbleSort(A)
for k <- 1 to n - 1
{
    flag <- 0
    for i <- 0 to n - k - 1
    {

        {
            swap(A[i], A[i+1])

````python
def bubbleSort(A):
    for j in range(len(A)):
        flag = False
        for i in range(len(A)-j-1):
            if A[i] > A[i+1]:
                temp = A[i]
                A[i] = A[i+1]
                A[i+1] = temp
                flag = True
        if not flag:
            break
````

````python
vals = generateRandomNumbers(10)
print('Before sorting:', vals)
bubbleSort(vals)
print('After sorting:', vals)
````

####Сложность алгоритма

Рассматривая различные алгоритмы решения одной и той же задачи, полезно проанализировать, сколько вычислительных ресурсов они требуют (время выполнения и память), и выбрать наиболее эффективный. Конечно, надо договориться о том, какая модель вычислений используется. В модели следуюет учитывать следующие характеристики:

* Однопроцессорная vs многопроцессорная
* Скорость чтения/записи (RAM, диск)
* 32 bit vs 64 bit
* Размер входа

В качестве модели вычислений будем использовать однопроцессорную машину с произвольным доступом (RAM), не предусматривающую параллельного выполнения операций и требующую:

* 1 единицу времени для выполнения арифметических и логических операций
* 1 единицу времени для выполнения присваивания и оператора return

Для сортировки методом пузырька предположим, что перестановка элементов занимает константное время $C$, тогда $T(n) = (n-1)(n-2)C = Cn^2 + 2Cn + 1 = O(n^2)$

Анализ сложности алгоритмов проводим для:

* Очень большого размера входа, т.е. при $n \rightarrow \infty$.
* Для самого плохого входа - худший случай (также рассматривают лучший случай и средний случай).

Кроме того, следует придерживаться следующих правил:

* Отбрасываем термы с наименьшим порядком.
* Отбрасываем множители.
* Время выполнения программы равно сумме времени выполнения всех ее фрагментов.


####О времени работы в худшем случае

Время работы в худшем случае определяется как максимальное время работы для входов данного размера:

* Зная время работы в худшем случае, мы можем гарантировать, что выполнение алгоритма закончится за некоторое время, даже не зная, какой именно вход попадется.
* На практие "плохие" входы могут попадаться достаточно часто.
* Время работы в среднем может быть довольно близко к времени работы в худшем случае.

````python
def linearSearch(A, n):
    for (index, element) in enumerate(A):
        if element == n:
            return index
    return -1

print(linearSearch([1,2,3,4,5], 2))
print(linearSearch([1,2,3,4,5], 6)) # Worst case
````

````python
def plotTimeComplexity(sortAlgorithm, max_size):
    import random
    import time
    import matplotlib.pyplot as plt
    
    times = []
    sizes = []
    
    for size in range(max_size):
        vals = generateRandomNumbers(size)
        start_time = time.time()
        sortAlgorithm(vals)
        stop_time = time.time()
        times.append(stop_time - start_time)
        sizes.append(size)
    
    plt.plot(sizes, times)
    plt.show()
````

````python
plotTimeComplexity(bubbleSort, 200)
````

####Алгоритм на псевдоязыке
```InsertionSort(A)
for i <- 1 to n-1
{
    value <- A[i] #c1 (n-1)
    hole <- i
    while hole > 0 and A[hole - 1] > value
    {
        A[hole] <- A[hole-1] #c2 
        hole <- hole - 1
    }
    A[hole] <- value #c3 (n-1)
}
```

####Сложность алгоритма

* Лучший случай, когда все значения упорядочены $T(n)  = (c_1+c_3)(n-1) = an + b = O(n)$
* Худший случай, когда все значения упорядочены, но в обратном порядке $T(n) = (c_1+c_3)(n-1) + (1 + 2 + 3 +...+ n-1)c_2 = (c_1+c_3)(n-1) + \frac{n(n-1)}{2}c_2 = an^2 + bn + c = O(n^2)$

````python
def insertionSort(A):
    for i in range(1, len(A)):
        temp = A[i]
        k = i
        while k > 0 and temp < A[k-1]:
            A[k] = A[k-1]
            k -= 1
        A[k] = temp
````

````python
vals = generateRandomNumbers(10)
print('Before sorting:', vals)
insertionSort(vals)
print('After sorting:', vals)
````

####Алгоритм на псевдоязыке
```
for i <- 0 to n - 2
{
    least <- i # c1 (n-1)
    for j <- i + 1 to n - 1
    {
        if A[j] < A[least] #c2 (n-1)+(n-2)+(n-3)+..+1 = n(n-1)/2
            least <- j
    }
    temp <- A[i] #c3 (n-1)
    A[i] <- A[least]
    A[least] <- temp
}
```

Временная сложность алгоритма:
$T(n) = (n-1)c1 + \frac{n(n-1)}{2}c2 + (n-1)c3 = an^2 + bn + c = O(n^2)$

````python
def selectionSort(A):
    for i in range(len(A)):
        least = i
        for k in range(i+1, len(A)):
            if A[k] < A[least]:
                least = k
        temp = A[least]
        A[least] = A[i]
        A[i] = temp
````

````python
vals = generateRandomNumbers(10)
print('Before sorting:', vals)
selectionSort(vals)
print('After sorting:', vals)
````

#Рекурсия (метод разделяй и властвуй)

Многие алгоритмы по природе **рекурсивны**: решая некоторую задачу они вызывают самих себя для решения ее подзадач. Идея метода "разделяй и властвуй" состоит как раз в этом. Сначала задача разбивается на несколько подзадач меньшего размера. Затем эти задачи решаются (с помощью рекурсивнго вызова - или непосредственно, если размер решаемой задачи стал достаточно мал). Наконец, их решения комбинируются и получается решение исходной задачи.

```def foo(параметры):
    if базовое условие для завершения рекурсии (задача стала достаточно малой):
        выполнение необходимых действий
        return некоторое значение
    else:
        выполнение необходимых действий
        рекурсивный вызов foo(измененные параметры)```

````python
def factorial(n):
    if n <= 1:
        return 1
    else:
        return n * factorial(n-1)
````

Давайте проанализируем этот алгоритм. Начнем с базового случая, когда размер входа равен 1:

$T(1) = 1$

Когда рамзмера входа равен n + 1, алгоритм тратит 1 единицу времени на умножение (или другую операцию), а затем происходит вызов такой же функции с размером входа n. Таким образом,

$T(n + 1) = T(n) + 1$

Если разложить этот терм, то получим:

$T(1) = 1$

$T(2) = T(1) + 1 = 2$

$T(3) = T(2) + 1 = 3$

$T(4) = T(3) + 1 = 4$

$...$

$T(n) = n$

Получается, что этот алгоритм тратит $n$ единиц времени на свою работу (т.е. $T(n) = n$).

````python
def factorial(n, acc=1):
    if n <= 1:
        return acc
    else:
        return factorial(n - 1, n * acc)
````

###Ханойская башня
В большом храме города Бенареса под куполом, накрывающим «центр вселенной», лежит медная плита, в которую вставлены три алмазные иглы длиной в локоть и толщиной с осиную талию. На одну из этих игл бог при сотворении мира надел шестьдесят четыре диска из чистого золота — самый большой из них лежит в самом низу на медной плите, и каждый диск, лежащий выше, меньше предыдущего. 

День и ночь священнослужители неустанно переносят диски с одной алмазной иглы на другую, руководствуясь навеки установленными и непреложными законами Брамы, по которым священнослужитель не должен двигать зараз более одного диска и всегда должен так переносить этот диск на иглу, чтобы под ним не оказалось диска, меньше его. Когда же, наконец, все 64 диска будут таким образом перенесены с той иглы, на которую бог поместил их при сотворении мира, на одну из двух других, то и башня, и храм, и сами брамины обратятся в прах; грянет гром, и мир исчезнет.

Число переносов, которые должны совершить брамины, чтобы перенести всю башню, равно $2^{64}-1$, а это потребует так много времени, что когда прозвучит последний, завершающий удар грома, мир будет в тысячу раз старше, чем сейчас.

````python
def brahma(n, a, b, c):
    if n != 0:
        brahma(n - 1, a, c, b)
        print('Moving from {} to {}'.format(a, c))
        brahma(n - 1, b, a, c)
````
